```python
from datetime import datetime as dt






def data_date(date):
    if date == '':
        return None
    else:
        return dt.strptime(date, "%Y-%m-%d")

def data_int(i):
    if i == '':
        return None
    else:
        return int(i)

for i in obj1:
    
    i['cancel_date'] = data_date(i['cancel_date'])
    i['days_to_cancel'] = data_int(i['days_to_cancel'])
    i['is_canceled'] = i['is_canceled'] == True
    i['is_udacity'] = i['is_udacity'] == True
    i['join_date'] = data_date(i['join_date'])
    
obj1[0]

    
    
    
```
